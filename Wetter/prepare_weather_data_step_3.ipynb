<h2># 2. Windstrke</h2>
<hr>

```python
df_wetter_wnd_sr_1.head()
```

```python
# WND_TC
'''
C = Calm
N = Normal
V = Variable
9 = Missing
(If a value of 9 appears with a wind speed of 0000, this indicates calm winds.)
'''
df_wetter_wnd_sr_1.groupby(['WND_TC'], as_index=True).agg({
    'WND_TC': 'count'
})
```

```python
# WND_SR_QC
'''
1 = Passed all quality control checks
5 = Passed all quality control checks, data originate from an NCEI data source
9 = Passed gross limits check if element is present
'''
df_wetter_wnd_sr_1.groupby(['WND_SR_QC'],as_index=True).agg({
    'WND_SR_QC' : 'count'    
})
```

```python
# WND_SR 
'''
A general domain comprised of the numeric characters (0-9). 9999 = Missing.
'''
df_wetter_wnd_sr_1.loc[(df_wetter_wnd_sr_1.WND_SR == 999.9)].groupby(['WND_SR']).agg({
    'WND_SR':'count'
})
```

```python
df_wetter_wnd_sr_1.loc[(df_wetter_wnd_sr_1.WND_SR == 0.0) & (df_wetter_wnd_sr_1.WND_TC == 9)].groupby(['WND_SR']).agg({
    'WND_SR':'count'
})
```

```python
# Dataframe ohne Missing values erstellen

df_wetter_wnd_sr_2 = df_wetter_wnd_sr_1.loc[(df_wetter_wnd_sr_1.WND_SR != 999.9)].copy()
df_wetter_wnd_sr_2.head()
```

```python
df_wetter_wnd_sr_2.groupby(['WND_TC'], as_index=True).agg({
    'WND_TC': 'count'
})
```

```python
df_wetter_wnd_sr_2.loc[(df_wetter_wnd_sr_1.WND_SR == 999.9)].groupby(['WND_SR']).agg({
    'WND_SR':'count'
})
```

```python
df_wetter_wnd_sr_2_grp = df_wetter_wnd_sr_2.groupby(['Jahr','Ort','Datum'], as_index=False, group_keys=False).agg({
    'WND_SR' : ['mean','min','max','median']
}).copy()
```

```python
# Anzahl prfen: 2007 bis 2008 ohne London; 2021 ohne NewYork;
df_wetter_wnd_sr_2_grp.groupby(['Jahr']).count()
```

```python
df_wetter_wnd_sr_2_grp.columns = ['_'.join(col) for col in df_wetter_wnd_sr_2_grp.columns.values]
```

```python
df_wetter_wnd_sr_2_grp.head()
```

```python
df_wetter_wnd_sr_2_grp.rename(columns={
    'Jahr_':'Jahr','Ort_':'Ort','Datum_':'Datum'
    ,'WND_SR_mean':'WND_SR_MEAN','WND_SR_min':'WND_SR_MIN','WND_SR_max':'WND_SR_MAX','WND_SR_median':'WND_SR_MEDIAN'}, inplace=True)
df_wetter_wnd_sr_2_grp.head()
```
