```python
df_wetter_wnd_tmp_dew_1.head()
```

```python
df_wetter_wnd_tmp_dew_1_grp = df_wetter_wnd_tmp_dew_1.groupby(['Jahr','Ort','Datum'], as_index=False, group_keys=False).agg({
    'TMP_AT' : ['mean','min','max','median'],
    'DEW_PT' : ['mean','min','max','median']
}).copy()
```

```python
df_wetter_wnd_tmp_dew_1_grp.head()
```

```python
df_wetter_wnd_tmp_dew_1_grp.columns = ['_'.join(col) for col in df_wetter_wnd_tmp_dew_1_grp.columns.values]
```

```python
df_wetter_wnd_tmp_dew_1_grp.head()
```

```python
df_wetter_wnd_tmp_dew_1_grp.rename(columns={
    'Jahr_':'Jahr','Ort_':'Ort','Datum_':'Datum'
    ,'TMP_AT_mean':'TMP_MEAN','TMP_AT_min':'TMP_MIN','TMP_AT_max':'TMP_MAX','TMP_AT_median':'TMP_MEDIAN'
    ,'DEW_PT_mean':'DEW_MEAN','DEW_PT_min':'DEW_MIN','DEW_PT_max':'DEW_MAX','DEW_PT_median':'DEW_MEDIAN'}, inplace=True)
df_wetter_wnd_tmp_dew_1_grp.head()
```
